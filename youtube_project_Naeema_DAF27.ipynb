# Data-Cleaning&Preprocessing

Data cleaning is an essential step in data analysis. Real-world datasets
often contain missing values, duplicate records, inconsistent data types,
or unnecessary columns.

In this section, we prepare the dataset for analysis by:

- Converting date columns to proper datetime format
- Filtering videos published from 2024 onwards
- Handling missing values
- Removing duplicate records
- Renaming incorrectly labeled columns

### Convert Date Columns

The dataset contains date-related columns stored as text. To perform
time-based analysis, we convert these columns into datetime format.

```python
df['publish_date'] = pd.to_datetime(df['publish_date'])
df['snapshot_date'] = pd.to_datetime(df['snapshot_date'])

df[['publish_date','snapshot_date']].head()
```

### Filter Data from 2024 Onwards

To focus on recent trends in YouTube content, we filter the dataset
to include only videos published from the year 2024 onwards.

```python
df = df[df['publish_date'].dt.year >= 2024]

df['publish_date'].dt.year.value_counts()
```

### Check Missing Values

Before analysis, we check whether the dataset contains missing values
in any columns.

```python
df.isnull().sum()
```

### Handle Missing Values

Some columns contain missing values.

Instead of removing rows, we replace missing values in  
**description**, **video_tags**, and **language** with `"Unknown"`.

```python
# Fill missing values
df['description'] = df['description'].fillna("Unknown")
df['video_tags'] = df['video_tags'].fillna("Unknown")
df['langauge'] = df['langauge'].fillna("Unknown")
```

```python
df.isnull().sum()
```

### Remove Duplicates

Duplicate records can affect analysis results.  
We check and remove duplicate rows if present.

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

### Rename Column

The column **langauge** is renamed to **language** for clarity.

```python
df.rename(columns={'langauge':'language'}, inplace=True)
```

```python
df.columns
```

# Feature-Engineering
<a id="feature-engineering"></a>

A new metric **engagement_rate** is created to measure audience interaction
based on likes and comments relative to views.

```python
df['engagement_rate'] = (df['like_count'] + df['comment_count']) / df['view_count']

df[['view_count','like_count','comment_count','engagement_rate']].head()
```

# Exploratory-Data-Analysis

Exploratory Data Analysis (EDA) helps uncover patterns, trends,
and relationships within the dataset.

In this section, we analyze key metrics such as **views, likes,
comments, rankings, and engagement rates** to understand how
YouTube videos perform across different countries and channels.

### Distribution of View Counts

This visualization shows how view counts are distributed among
trending YouTube videos. It helps identify whether most videos
receive moderate views or if a few videos dominate with extremely
high view counts.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['view_count'], bins=50)

plt.title("Distribution of Video View Counts")
plt.xlabel("View Count")
plt.ylabel("Frequency")

plt.show()
```

**Insight**

The distribution of view counts is highly right-skewed.  
Most trending videos receive moderate views, while a small number of videos achieve extremely high view counts.
A few videos become very viral, while most videos get normal or moderate views.indicating that viral videos are relatively rare.

### Distribution of Like Counts

This chart shows how likes are distributed across trending videos.
It provides insight into how audiences interact with popular content.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['like_count'], bins=50)

plt.title("Distribution of Like Counts")
plt.xlabel("Like Count")
plt.ylabel("Frequency")

plt.show()
```

**Insight**

Like counts follow a similar pattern to views, where most videos receive moderate likes while only a few videos generate very high audience engagement.

### Relationship Between Views and Likes

This scatter plot examines the relationship between video views
and likes. It helps determine whether videos with more views
also tend to receive more likes.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='view_count', y='like_count', data=df)

plt.title("Views vs Likes")
plt.xlabel("View Count")
plt.ylabel("Like Count")

plt.show()
```

**Insight**

There is a strong positive relationship between views and likes.  
Videos with higher view counts tend to receive more likes.
As views increase, likes also increase.
Popular videos tend to get more likes.

### Relationship Between Views and Comments

This visualization explores whether highly viewed videos
also generate more comments from viewers.

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='view_count', y='comment_count', data=df)

plt.title("Views vs Comments")
plt.xlabel("View Count")
plt.ylabel("Comment Count")

plt.show()
```

**Insight**

Videos with higher view counts generally receive more comments, indicating increased audience interaction for popular videos.
More viewers → more people comment.
Popular videos create more discussion.

### Top Countries with Trending Videos

This chart shows which countries appear most frequently
in the trending videos dataset.

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)

plt.title("Top 10 Countries with Trending Videos")
plt.xlabel("Number of Trending Videos")
plt.ylabel("Country")

plt.show()
```

**Insight**

A small number of countries dominate the trending video dataset, indicating higher content production and viewer engagement in those regions.
Some countries produce more trending videos.
Those countries have larger YouTube activity or audience.

### Average View Count by Country

To understand how video popularity varies across different regions,
we calculate the **average view count for each country** using the
groupby operation.

This helps identify which countries tend to have higher average
views for trending videos.

```python
# Average views by country

country_views = df.groupby('country')['view_count'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=country_views.values, y=country_views.index)

plt.title("Top Countries by Average Video Views")
plt.xlabel("Average View Count")
plt.ylabel("Country")

plt.show()
```

**Insight**

The chart shows the countries with the highest average video views.
This indicates that trending videos in these countries tend to
receive higher viewer engagement.

### Top Channels by Total Views

This visualization identifies channels that accumulate
the highest total views among trending videos.

```python
top_channels = df.groupby('channel_name')['view_count'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_channels.values, y=top_channels.index)

plt.title("Top Channels by Total Views")
plt.xlabel("Total Views")
plt.ylabel("Channel")

plt.show()
```

**Insight**

A few channels accumulate significantly higher total views than others, suggesting strong audience reach and consistent content popularity.
Some channels consistently produce very popular videos.
These creators have strong audience reach.

### Engagement Rate Distribution

This chart shows how audience engagement varies across
trending videos based on likes and comments relative to views.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['engagement_rate'], bins=50)

plt.title("Distribution of Engagement Rate")
plt.xlabel("Engagement Rate")
plt.ylabel("Frequency")

plt.show()
```

**Insight**

Some videos make viewers more active.Most videos have relatively low engagement rates, while a small number of videos achieve higher audience interaction.

### Distribution of Daily Rank

This visualization shows how videos are distributed
across different trending ranks.

```python
plt.figure(figsize=(10,6))
sns.histplot(df['daily_rank'], bins=50)

plt.title("Distribution of Daily Rank")
plt.xlabel("Daily Rank")
plt.ylabel("Frequency")

plt.show()
```

**Insight**

Daily Rank-position of a video in the trending list.The distribution shows that many videos cluster around certain ranking ranges, reflecting how trending videos are distributed within the ranking system.

### Correlation Heatmap

A heatmap shows how strongly different variables are related to each other.Here,
the heatmap shows relationships between numerical variables
such as views, likes, comments, and engagement rate.

```python
plt.figure(figsize=(8,6))

corr = df[['view_count','like_count','comment_count','engagement_rate']].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

**Insight**

Views, likes, and comments show strong positive correlations, indicating that highly viewed videos tend to receive more likes and comments.
Views ↑ → Likes ↑
Views ↑ → Comments ↑
That means they have positive correlation.

### Video Publishing Trend

It shows how many trending videos were published on different dates.

```python
publish_trend = df['publish_date'].dt.date.value_counts().sort_index()

plt.figure(figsize=(12,6))
publish_trend.plot()

plt.title("Trending Video Publishing Trend")
plt.xlabel("Date")
plt.ylabel("Number of Videos")

plt.show()
```

**Insight**

Some days:
More videos appear

Some days:
Fewer videos appear

The publishing trend shows fluctuations in the number of trending videos over time, highlighting variations in video release activity.
